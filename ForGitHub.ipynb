```python
from github import Github
import os

# Данные аутентификации GitHub
username = "your_username"
password = "your_password"  # Можно использовать токен доступа вместо пароля

# Имя репозитория и владельца
repository_owner = "repository_owner_username"
repository_name = "repository_name"

# Путь к файлу, который нужно загрузить
file_path = "path_to_file"

# Инициализация экземпляра PyGithub
g = Github(username, password)

# Получение репозитория
repo = g.get_user(repository_owner).get_repo(repository_name)

# Создание коммита для загрузки файла
commit_message = "Загрузка файла {}".format(os.path.basename(file_path))
repo.create_file(os.path.basename(file_path), commit_message, open(file_path, "rb").read())

print("Файл успешно загружен в репозиторий на GitHub!")
```
